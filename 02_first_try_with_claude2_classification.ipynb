**In this notebook, I ask the Generative AI, Claude 2, to predict a topic and a verbal contribution for each sentence, and then compare these with the predefined categories to assess the model's accuracy.**

# Packages

```python
import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

# Topic

## Data

```python
df = pd.read_excel('/Users/ningyuhan/Desktop/Capstone_files/first_try_claude_2.xlsx')
```

```python
df.head(5)
```

## Accuracy visualization

```python
cm = confusion_matrix(df['Actual'], df['Predict(claude_2)'])

plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=sorted(df['Actual'].unique()), yticklabels=sorted(df['Predict(claude_2)'].unique()))
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
accuracy = accuracy_score(df['Actual'], df['Predict(claude_2)'])
print("Accuracy:", accuracy)
```

# Verbal Contribution - 4h

```python
df_1 = pd.read_excel('/Users/ningyuhan/Desktop/Capstone_files/first_try_4h.xlsx')
```

```python
df_1.head(5)
```

```python
df_1['actual_numeric'] = df_1['actual'].map({'4h': 1, 0: 0})
df_1['predict_numeric'] = df_1['predict'].map({'4h': 1, 0: 0})
```

```python
cm = confusion_matrix(df_1['actual_numeric'], df_1['predict_numeric'])
```

```python
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=sorted(df_1['actual_numeric'].unique()), yticklabels=sorted(df_1['predict_numeric'].unique()))
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
accuracy = accuracy_score(df_1['actual_numeric'], df_1['predict_numeric'])
print("Accuracy:", accuracy)
```
